```python
cleaned_desc = []
for desc in productDescription.Description:
    filtered_text = []
    for words in desc.split():
        if not re.findall(r'[^\w\s]', words.strip()):
            filtered_text.append(words)
    cleaned_desc.append(' '.join(filtered_text))
```

```python
len(cleaned_desc)
```

```python
cleaned_title = []
for desc in productDescription.Name:
    filtered_text = []
    for words in desc.split():
        if not re.findall(r'[^\w\s]', words.strip()):
            filtered_text.append(words)
    cleaned_title.append(' '.join(filtered_text))
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
print(len(stop_words))
```

```python
stop_words.add('get')
```

```python
filtered=[]
for w in tokenize_desc_word:
     if w not in stop_words:
        filtered.append(w)
print("Tokenized Sentence:",tokenize_desc_word, '\n\n')
print("Filterd Sentence:",filtered)
```

```python
fdist = FreqDist(filtered)
print(len(fdist))
```

```python
fdist.most_common(10)
```

```python
import matplotlib.pyplot as plt
fdist.plot(10, cumulative=False) # The number 10 here plots the top 10 words only
```
